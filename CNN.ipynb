```python
from keras.utils import plot_model
from keras.models import Model
from keras.layers import Input
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
```

```python
'''
the model is a cnn image classifier :

input                         conv+pool                      conv+pool                              full connection

input : (64,64,1) = 64,64 sized image of 1 color channel (b/w)

full connection requires flattening the pooling output

'''
```

```python
visible = Input(shape=(64,64,1))
conv1 = Conv2D(32, (4,4), activation = 'relu')(visible)
pool1 = MaxPooling2D()(conv1)
conv2 = Conv2D(16, (4,4), activation = 'relu')(pool1)
pool2 = MaxPooling2D()(conv2)
flatten = Flatten()(pool2)
hidden = Dense(10, activation = 'relu')(flatten)
out = Dense(1, activation = 'sigmoid')(hidden)
```

```python
model = Model(inputs = visible, outputs = out)
```

```python
model.summary()
```

```python
plot_model(model)
```
