```python
'''
We're interested in the G, so we take the saved model (in DCGAN CIFAR Saved files)

and we use it to generate new images.

note that the pixel values generated are between -1,1 because of our implementation of training
in DCGAN CIFAR

and so we need to scale then to plot them.
'''
```

```python
from keras.models import load_model
from numpy.random import randn
import matplotlib.pyplot as plt
```

```python
def generate_latent_points(n_samples, latent_dims):
    x = randn(n_samples * latent_dims)
    x = x.reshape(n_samples, latent_dims)

    return x
```

```python
def save_plot(samples, n):

    #samples = (samples + 1) / 2.0

    for i in range(n * n):

        plt.subplot(n, n, i+1)
        plt.axis('off')
        plt.imshow(samples[i, :, :]) # ith sample, all rows, all cols
    plt.show()
```

```python
Gen = load_model('gen_3.h5')
```

```python
latent_in = generate_latent_points(100, 100)
X = Gen.predict(latent_in)

X = (X + 1) / 2.0

save_plot(X, 10)
```
