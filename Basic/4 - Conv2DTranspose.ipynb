```python
from numpy import asarray
from keras.models import Sequential
from keras.layers import Conv2DTranspose
```

```python
'''
Conv2DTranspose takes args : (number of  filters, kernel size, strides, input_shape)

here strides refer to output strides
that means if on one element say [1] the layer uses the stride of (2x2) the output is a 2x2 array:
 e.g : with w = 1 and b = 0:
                 1 ----- conv2dtrans( strides = (2,2) -------> [[1,0],
                                                                [0,0]]

i.e the element gets upsampled with the surrounding adjusted by a calculation using weights
'''
```

```python
model = Sequential([
    Conv2DTranspose(1, (1,1), strides = (2,2), input_shape = (2,2,1))
])
```

```python
weights = [asarray([[[[1]]]]), asarray([0])]
model.set_weights(weights)
```

```python
'''

i want see what is happening
so im generating a random 2x2 array with gaussian values
and plotting it and the models output for it

'''
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a 2x2 array of random Gaussian elements
X = np.random.normal(0, 1, (2, 2))

# Plot the array as a color image
plt.imshow(X, cmap='viridis', interpolation='nearest')
plt.colorbar()  # Add a colorbar to the plot
plt.title('Random 2x2 Array')
plt.show()
```

```python
X = X.reshape(1,2,2,1)
```

```python
y_hat = model.predict(X)
```

```python
y_hat = y_hat.reshape((4,4))
```

```python
plt.imshow(y_hat, cmap='viridis', interpolation='nearest')
plt.colorbar()  # Add a colorbar to the plot
plt.title('ConvTransposed')
plt.show()
```
