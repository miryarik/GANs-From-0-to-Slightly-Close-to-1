```python
from keras.layers import Conv2DTranspose
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import Dense
from keras.layers import Reshape
from keras.utils import plot_model
from numpy import random
from numpy import asarray
import matplotlib.pyplot as plt
```

```python
# seed random number generator
random.seed(random.randint(0,100))
X = random.normal(0, 1, (100, 1))
X = X.reshape(1,100)
```

```python
X_plot = X.reshape(10,10)
```

```python
plt.imshow(X_plot, cmap='viridis')
plt.colorbar()  # Add a colorbar to the plot
plt.title('Random 10x10 noise')
plt.show()
```

```python
print(X)
```

```python
model = Sequential([
    Dense(128*5*5, input_dim = 100),
    Reshape((5,5,128)),
    Conv2DTranspose(1, (3,3), strides = (2,2), padding = 'same')
])
```

```python
model.summary()
```

```python
y_hat = model.predict(X)
```

```python
y_hat = y_hat.reshape((10,10))
```

```python
plt.imshow(y_hat, cmap='viridis')
plt.colorbar()  # Add a colorbar to the plot
plt.title('Generated (10,10) Array')
plt.show()
```
