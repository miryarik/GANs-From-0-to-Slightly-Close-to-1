```python
from keras.layers import UpSampling2D
from keras.layers import Dense
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import Reshape
from keras.utils import plot_model
```

```python
model = Sequential([
    Dense(128*5*5, input_dim = 100),
    Reshape((5,5,128)),
    UpSampling2D(),
    Conv2D(1, (3,3), padding = 'same')
])
```

```python
model.summary()
```

```python
plot_model(model, show_shapes=True, show_layer_names=True)
```

```python
'''
i want to see it work so im gonna:

1 - give it a gaussian latent space of (100,1)
2 - predict directly with no training
3 - plot the result as a b/w scatter

'''
```

```python
# generate random Gaussian values
from numpy import random
from numpy import ndarray
```

```python
# seed random number generator
random.seed(random.randint(0,100))
input_data = random.normal(0, 1, (100, 1))
input_data = input_data.reshape(1,100)
```

```python
print(input_data)
```

```python
out = model.predict(input_data)
```

```python
import matplotlib.pyplot as plt
plt.imshow(out[0, :, :, 0], cmap='gray')
plt.show()
```
