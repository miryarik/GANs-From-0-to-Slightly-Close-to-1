```python
from keras.layers import UpSampling2D
from keras.models import Sequential
from numpy import asarray
```

```python
'''
check usage for UpSampling2D

optional params :

size(2-tuple): default (2,2)
            it takes a param 'size' as (rows,columns,channels) OF THE DESIRED OUTPUT (not input)
            (2,2) is the default and means x2 all the rows and columns
            (2,3) would mean x2 the rows and x3 the columns

interpolation(string): default 'nearest'
            this takes the name of the neighbour finding algorithm.
            the default is 'nearest' referring to nearest-neighbour algo
            another is 'bilinear'

'''
```

```python
model = Sequential([
    UpSampling2D(input_shape = (2,2,1), size = (2,3), interpolation = 'bilinear')
])
```

```python
'''
i wanted to see what is happening so i used :

            input X = a random (2,2) array as a color plot

            note that the kernel size is (2,3) so,

            the output is a (4,6) upsampled array
'''
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a 2x2 array of random Gaussian elements
X = np.random.normal(0, 1, (2, 2))

# Plot the array as a black and white image
plt.imshow(X, cmap='viridis')
plt.colorbar()  # Add a colorbar to the plot
plt.title('Random 2x2 Array')
plt.show()
```

```python
X = X.reshape((1,2,2,1))
```

```python
model.summary()
```

```python
y_hat = model.predict(X)
print(y_hat.shape)
```

```python
y_hat = y_hat.reshape((4,6))
```

```python
plt.imshow(y_hat, cmap='viridis')
plt.colorbar()  # Add a colorbar to the plot
plt.title('UpSampled (4,4 Array)')
plt.show()
```
